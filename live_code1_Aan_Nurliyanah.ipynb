c. Hitunglah rata-rata nilai kapasitas untuk arus = 1 Ampere dan cycle ganjil.

```python
rata_rata= df_hambatan[
    (df_hambatan['kuat_arus'] == 1) &
    (df_hambatan['siklus'] % 2 == 1 ) 
].groupby(['kuat_arus','siklus'])['kapasitas_baterai'].mean().sort_values(ascending=False)
```

Menghitung nilai kapasitas menggunakan group by, dimana kuat_arus = 1 dan siklus mod 2 = 1 ,untuk mencari nilai ganjilnya. 

diurutkan berdasarkan mean dari kecil ke besar.
